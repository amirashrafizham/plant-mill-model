```python
# Load the Excel file data in sheet number 2 using pandas
import pandas as pd
pd.set_option("display.max_columns", None)

# Load the Excel file data in sheet number 2 using pandas, use the first row as the column names
df = pd.read_excel("Case Study for Data Scientist.xlsx", sheet_name="Sheet1", header=0)
df.head()

categorical_features = df.select_dtypes(include=["object"]).columns
print(categorical_features)
print(len(categorical_features))

numerical_features = df.select_dtypes(include=["int64", "float64"]).columns
# numerical_features = numerical_features.drop('Actual OER %')
print(numerical_features)
print(len(numerical_features))

label = "Actual OER %"
```

```python
# ZO22 is chosen as the mill code to be analyzed
zo22 = df[df["Mill Code"] == "Z013"]
# set the Date as the index
zo22.head(10)
```

```python
# From the Date column filter the Data from 2024-12-01 onwards
zo22_df = zo22[zo22["Date"] >= "2024-10-01"]
zo22_df.head(10)
# Remove the categorical columns
zo22_df = zo22_df.drop(columns=categorical_features)
# Remove the null values
zo22_df = zo22_df.dropna()
# Set the date as the index
zo22_df = zo22_df.set_index("Date")

refined_features = numerical_features.drop(["Other Seeds %", "Seed B %", "Inland %"])
# Remove the columns that are not required from zo22_df based on the refined_features
order = [
        "Crop Freshness Score",
        "Ripe %",
        "Long Stalk %",
        "Rat Damage %",
        "Loose Fruits %",
        "Rainfall (mm)",
        "Age Profile (years)",
        "Total Oil Losses %",
        "Downtime %",
        "FFB Processed (MT)",
        "Seed A %",
        "Coastal %",
    ]
zo22_df = zo22_df[order]
```

```python
zo22_df.head(10)
```

```python
# Load the model 
from joblib import load

model = load("zir_trained.pkl")
# Predict the Actual OER % using the model
zo22_df["Predicted OER %"] = model.predict(zo22_df)
zo22_df.head(10)
```

```python
import warnings
warnings.filterwarnings("ignore")
from scipy.optimize import differential_evolution


# Define the bounds for the features

bounds = [
    (280, zo22["Crop Freshness Score"].max()),  # Crop Freshness Score
    (90, zo22["Ripe %"].max()),  # Ripe %
    (zo22["Long Stalk %"].min(), 5),  # Long Stalk %
    (zo22["Rat Damage %"].min(), 5),  # Rat Damage %
    (8, zo22["Loose Fruits %"].max()),  # Loose Fruits %
    (zo22["Rainfall (mm)"].min(0), 5),  # Rainfall (mm)
    (13.92, 13.92),  # Age Profile (years)
    (1.92, 1.92),  # Total Oil Losses %
    (0.06, 0.06),  # Downtime %
    (0, zo22['FFB Processed (MT)'].max()),  # FFB Processed (MT)
    (0.44, 0.44),  # Seed A %
    (63, 63),  # Coastal %
]
#print(zo22["Crop Freshness Score"].max())
#print(zo22["Ripe %"].max())
#print(zo22["Ripe %"].min())
#print(zo22["Long Stalk %"].min())
#print(zo22["Rat Damage %"].min())
#print(zo22["Loose Fruits %"].max())
#print(zo22["Rainfall (mm)"].min())
#print(zo22["FFB Processed (MT)"].max())


zo22_df_day1 = zo22_df[zo22_df.index == "2024-10-01"]
zo22_df_day1.head()

def obj_fun(X):
    X = [X]
    results = model.predict(X)
    return -results


# Optimize the objective function using differential_evolution
result = differential_evolution(obj_fun, bounds)

# Print the optimized values

print(f"Optimized value for Crop Freshness Score: {result.x[0]}")
print(f"Optimized value for Ripe %: {result.x[1]}")
print(f"Optimized value for Long Stalk %: {result.x[2]}")
print(f"Optimized value for Rat Damage %: {result.x[3]}")
print(f"Optimized value for Loose Fruits %: {result.x[4]}")
print(f"Optimized value for Rainfall (mm): {result.x[5]}")
print(f"Optimized value for Age Profile (years): {result.x[6]}")
print(f"Optimized value for Total Oil Losses %: {result.x[7]}")
print(f"Optimized value for Downtime %: {result.x[8]}")
print(f"Optimized value for FFB Processed (MT): {result.x[9]}")
print(f"Optimized value for Seed A %: {result.x[10]}")
print(f"Optimized value for Coastal %: {result.x[11]}")
print(f"Optimized value for Actual OER %: {-result.fun}")
print(-result.fun)
```
